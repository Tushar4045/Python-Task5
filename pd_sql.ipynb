```python
import pandas as pd
```

```python
cust = pd.read_csv('customers.csv',header=0,index_col=0)
cust
```

```python
order = pd.read_csv('D:/Python practice/pythonTask/Python-Task5/orders.csv',header=0,index_col=0)
order
```

```python
cust_order = pd.merge(cust,order, on = 'CustomerID')
cust_order
```

```python
aggAmt = cust_order.agg(
    {
        'OrderAmount': ['sum','min','max'],
    }
)
aggAmt
```

```python
custgp = cust_order.groupby('CustomerName').agg({'OrderAmount': 'sum'}).sort_values('OrderAmount', ascending=False)
custgp
```

```python
citygp = cust_order.groupby('City').agg(
    {
        'OrderAmount': 'mean'
    }
)
citygp
```

```python
# cust_order.agg({'OrderAmount': 'sum'})
cust_order[cust_order['OrderAmount']>2000].agg({'OrderAmount': 'sum'})
```

```python
cust_city = cust_order.groupby('City').agg({'CustomerName': 'count'})
cust_city
```
